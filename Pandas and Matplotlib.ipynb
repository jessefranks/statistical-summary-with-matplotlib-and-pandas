11. Now consider a modified version of the bank data that contains missing values (bank_data_missing_vals.csv). (a) Using Pandas determine all the attributes with missing values and the number of missing values for each such attribute. (b) Show all the instances in the data that contain a missing value. (c) Fill the missing values for all numeric attributes using the mean value for the attribute. (d) After filling in the missing numeric values, drop all rows where a categorical attribute contains a missing value. (e) Show that the final resulting table does not contain missing values

```python
missing_df = pd.read_csv("C:\\Users\\Jesse\\Downloads\\!classes\\Depaul\\AU 20\\bank_data_missing_vals.csv", 
                         na_values=["?"])
```

```python
missing_df.info() # Using Pandas determine all the 
# attributes with missing values and the number of missing values for each such attribute

# 1   age           590 non-null    10 missing
# 2   income        587 non-null    13 missing
# 4   gender        590 non-null    10 missing
# 5   region        591 non-null    9 missing
```

```python
missing_df # looking at the df.
```

```python
# Show all the instances in the data that contain a missing value.
missing_df[missing_df.isnull().any(axis=1)] 
```

```python
#  1   age           590 non-null    float64
#  2   income        587 non-null    float64
#  4   gender        590 non-null    object 
#  5   region        591 non-null    object 

#  Fill the missing values for all numeric attributes using the mean value for the attribute.
age_mean = missing_df.age.mean()
missing_df.age.fillna(age_mean, axis=0, inplace=True)

income_mean = missing_df.income.mean()
missing_df.income.fillna(income_mean, axis=0, inplace=True)
```

```python
missing_df.info() # age and income are filled in with the means.
```

```python
# After filling in the missing numeric values, drop all rows where a categorical attribute contains a missing value.
missing_df.drop(missing_df[missing_df.gender.isnull()].index, axis=0, inplace=True)
missing_df.drop(missing_df[missing_df.region.isnull()].index, axis=0, inplace=True)
```
